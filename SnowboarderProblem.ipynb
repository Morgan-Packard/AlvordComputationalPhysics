A snowboarder drops from a height *h* landing on a sloping surface. The speed of the snowboarder perpendicular to the slope (i.e. along y) when they first touch the surface is:<br> $$v_{y}=-\sqrt{2gh}\cos\theta.$$<br> During the landing the snowboarder drops their center of mass vertically by an amount *b* by bending their knees. The average acceleration perpendicular to slope during the landing is: <br> $$a_{N}=\frac{gh\cos\theta}{b}.$$<br> and the average normal force can be found using <br> $$F_{N}=mg(1+\frac{h}{b})\cos\theta.$$<br>  The duration for the time of the landing can also be derived and is found using <br> $$\Delta t=b\sqrt\frac{2}{gh}.$$<br>

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt

#Knowns - what variables did the problem give us? don't forget units!
mass = 72 # kg, mass of the snowboarder
drop_height = 3 # m, drop height
g_const = 9.8 # N/kg, gravity
bend = 0.4 # m, drop in center of mass during knee bend
theta = 0 # deg, angle of the slope


#Lists - these will typically be used to create graphs, so think about what we want to graph! Don't forget units
acceleration_data = [] # m/s
normal_force_data = [] # N
theta_data = [] #deg
#The Motion - Using a while loop, we update the various variables of the object

# 5997.6

for theta in range(91):
    A_N = g_const*drop_height*np.cos(np.deg2rad(theta))/bend # solve for a_N as defined in the problem description
    F_N = mass*g_const*(1+drop_height/bend)*np.cos(np.deg2rad(theta)) # solve for F_N as defined in the problem description

    # Insert the data into appropriate lists
    acceleration_data.append(A_N)
    normal_force_data.append(F_N)
    theta_data.append(theta)

    #for question 3, simple if statement to printout theta and F_N
    if theta == 60:
        print(theta, F_N)

#ploting
fig, plot = plt.subplots(2, figsize=(10,5)) # creates a plot with 2 subplots
fig.subplots_adjust(hspace=0.5) #adjusts the space between the plots

#details of the theta v acceleration scatter plot
plot[0].set(xlabel="Theta (degrees)", ylabel="Acceleration (m/s)", title="Acceleratation vs Theta") 
plot[0].scatter(theta_data, acceleration_data, color='pink', edgecolor='blue'); 

#details of the normal force v theta scatter plot
plot[1].set(xlabel="Theta (degrees)", ylabel="Normal Force (N)", title="Normal Force vs Theta") 
plot[1].scatter(theta_data, normal_force_data, color='pink', edgecolor='blue');   # this makes a scatterplot!

fig.show() 
```
