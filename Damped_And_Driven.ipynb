Challenge 0 - Velocity vs time graph

For this set, we are going to remove friction. The easiest way to do this is to set μ to 0. Don’t delete the force as we will add it back in later. Add in a velocity vs time graph to your springs code and then answer the following questions in a markdown box.

Questions to answer:
Compare and contrast the general shapes of your position vs time and velocity vs time graphs
Both are sinusoidal, but they are out of phase from one another and have differing amplitudes but share the same frequency.

Where is the object located when it is moving fastest? How do you know?
The object is located at the location x = 1 (where the spring would naturally be if unstretched)

Where is the object located when it is stopped? How do you know?
The object is located at either x = ~1.5 or ~0.5. This is because the velocity is 0 at these times.

What is the period (time to complete one cycle) of the motion as seen on the position vs time graph?
~1.7 seconds

What is the period (time to complete one cycle) of the motion as seen on the velocity vs time graph?
Comment on your answers to 4 and 5. 
~1.6 seconds. This aligns with the answers from 4 and 5.

```python
#import stuff
import matplotlib.pyplot as plt
import numpy as np

#define the object parameters
m = 2 #kg mass of block on horizontal surface 
k = 25 #N/m spring constant 
L = 1 #m unstretched length of spring 

#define simulation parameters
x = 0.5 #m initial position of the block 
v = 0 #m/s initial velocity of the block
t = 0 #s initial time
g = 9.8 # N/kg gravitational constant on Earth's surface
dt = 0.001 #s simulation time step
sim_length = 10 #s simulation length
mu_k = 0 # coeffeicient of kinetic friction

#initialize lists for graphing later
x_list=[x] #m
v_list=[v] #m/s
t_list=[t] #s

spring_en_list=[(0.5*k*(x-L)**2)] #N
kinetic_en_list=[(1/2 * m * v**2)] #N
total_en_list=[(0.5*k*(x-L)**2) + (1/2 * m * v**2)]

while t<sim_length: #this simulation will run for the length of time set by sim_length
    vmag = np.linalg.norm(v)

    if vmag == 0:
        F_friction = 0 # N, force of kinetic friction when ovject is stopped
    else:
        vhat = v/vmag #unit vector, holding the direction of the velocity.
        F_friction = - vhat * mu_k * m * g

    F_spring = -k * (x-L) #N, code Hooke's law here

    F_net = F_spring + F_friction #N, F_net lol
    a = F_net/m  # solving for acceleration with respect to N2L
    v = v + a*dt # solving for velocity with respect to acceleration
    x = x + v*dt # solving for position with respect to velocity
    t = t + dt  # iterate time
    
    #append lists
    x_list.append(x) #m
    v_list.append(v) #m/s
    t_list.append(t) #s
    spring_en_list.append((0.5*k*(x-L)**2))
    kinetic_en_list.append((1/2 * m * v**2))
    total_en_list.append((0.5*k*(x-L)**2) + (1/2 * m * v**2))

fig, plot = plt.subplots(2, figsize=(10,5)) # creates a plot
fig.subplots_adjust(hspace=0.5) #adjusts the space between the plots

#Graph of position vs time for horizontal oscillator
# plt.plot(t_list,spring_en_list)
# plt.plot(t_list,kinetic_en_list)
# plt.plot(t_list,total_en_list)
plot[0].set(xlabel="time (s)", ylabel="position (m)", title='Position vs Time oscillator')
plot[0].plot(t_list,x_list)
# plot[0].xlabel('time (s)')
# plot[0].ylabel('energy (J)')

plot[1].set(xlabel="time (s)", ylabel="velocity (m/s)", title='Velocity vs Time oscillator')
plot[1].plot(t_list,v_list)

plt.show()
```
