```python
#import proper libraries
import numpy as np
import matplotlib.pyplot as plt

#initialize values
x = float(input("What is the tower height")) #m, input tower height -> ball position
v = 0 #m/s, initial velocity
a = -9.8 # m/s/s, gravity
t = 0 # s, time
dt = 0.1 # s, change in time

#initialize lists to store data
xlist = [x] # position data
vlist = [v] # velocity data
tlist = [t] # time data

while (x > 0):
    v = v + a*dt  #adjusting velocity with respect to acceleration
    x = x + v*dt  #adjusting position with respect to velocity
    t = t + dt

    #appending to lists to store data from each iteration
    xlist.append(x) #xlist
    vlist.append(v) #vlist
    tlist.append(t) #tlist

print("the ball hit the ground at ", t, " seconds") # outputs the time when ball position <= 0

fig, ax = plt.subplots(2, figsize=(10,5)) # creates figure with 2 subplot
fig.subplots_adjust(hspace=0.5) #adjusts the space between the plots

ax[0].set(xlabel="Time (s)", ylabel="Position (m)", title="Position vs Time") # sets axis labels and title
ax[0].plot(tlist, xlist) # plots the x vs t line

ax[1].set(xlabel="Time (s)", ylabel="Velocity (m/s)", title="Velocity vs Time") # sets axis labels and title
ax[1].plot(tlist, vlist) # plots the v vs t line
```
