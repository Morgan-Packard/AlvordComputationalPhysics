```python
import matplotlib.pyplot as plt         # "plt" is much shorter
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5, 6])
y = [3,6,9,10,12,15,18]
y2 = [2,4,6,8,10,12,14]
m, b = np.polyfit(x, y, 1)
m2, b2 = np.polyfit(x, y2, 1)
y_model = m*x+b
y_model2 = m2*x+b2

plt.plot(x, y_model)
plt.plot(x, y_model2)
plt.text(1,16, "Memes: y=" + '{:.2f}'.format(b) + ' + {:.2f}'.format(m) + "x")
plt.text(1,15, "Memes: y=" + '{:.2f}'.format(b2) + ' + {:.2f}'.format(m2) + "x")

plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs y")

plt.scatter(x, y, color='pink', edgecolor='blue');   # this makes a scatterplot!
plt.scatter(x, y2, color='brown', edgecolor='purple');   # this makes a scatterplot!

plt.show() 
```
