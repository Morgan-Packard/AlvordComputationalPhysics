```python
# put the needed libraries first, here at the top
import matplotlib.pyplot as plt         # "plt" is much shorter
import numpy as np

#Knowns - what variables did the problem give us? don't forget units!
m = 1 # kg, mass
t = 0 #s, Starting time
g = np.array([0, -9.8, 0]) # N/kg gravitational force
pos = np.array([0, 0, 0]) #m, Initial position
v = np.array(np.sqrt([2, 2, 0])) #m/s, velocity
dt = 0.01 #s
Omega = np.array([0, 0, 0]) #spin
surfaceArea = 0.00502654825 # m^2, surface areahttps://www.dimensions.com/element/ping-pong-ball-table-tennis-ball#:~:text=Ping%2DPong%20Balls%2C%20or%20Table,4.94%E2%80%9D%20(125.6%20mm).
rho = 1.225 # kg/m^3, air density https://en.wikipedia.org/wiki/Density_of_air#:~:text=At%20101.325%20kPa%20(abs)%20and%2015%20%C2%B0C%20(59,International%20Standard%20Atmosphere%20(ISA). 
C_drag = 0.5 # Drag coefficient of a ping pong ball https://scienceworld.wolfram.com/physics/DragCoefficient.html 

#Lists - these will typically be used to create graphs, so think about what we want to graph! Don't forget units
tlist = [t] #s
xlist = [pos[0]] #m
ylist = [pos[1]] #m
vxlist = [v[0]] #m/s
vylist = [v[1]] #m/s

#The Motion - Using a while loop, we update the various variables of the object
while pos[1] >= 0: #example - while t < 10 - read the question closely. Under what condition do we want to the code to stop running?
    F_grav = m*g # gravitational force
    F_drag = 1/2 * C_drag * rho * surfaceArea * (v)**2
    F_lift = np.linalg.cross(Omega, v) 

    Fnet = F_grav + F_drag + F_lift # net force
    a = Fnet/m  #solve the acceleration
    v = v + a*dt  #solve the acceleration
    pos = pos + v*dt  #solve the acceleration
    t = t + dt  #solve the acceleration

    #we also need to update our lists
    tlist.append(t)
    xlist.append(pos[0])
    ylist.append(pos[1])
    vxlist.append(v[0])
    vylist.append(v[1])

plt.xlabel("x pos (m)")
plt.ylabel("y pos (m)")
plt.title("tragectory graph")

plt.scatter(xlist, ylist, color='pink', edgecolor='blue');   # this makes a scatterplot!

plt.show() 
```
