A Uranium-235 atom (m = 3.90 x 10^-25 kg) is ionized (meaning an electron has been removed). It is placed (meaning starting from rest) into a uniform electric field created by two vertical parallel plates that are d = 5 cm apart. If the electric field has a magnitude of 50,000 N/C, how fast is the particle moving when it leaves the other side? Code this scenario using Newton’s 2nd Law and the Euler Cromer Method and print out this final velocity. (code)

```python
#Part A
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for plot attributes
plt.rc("xtick", labelsize="large")
plt.rc("ytick", labelsize="large")
plt.rc("axes", labelsize="xx-large")
plt.rc("axes", titlesize="xx-large")
plt.rc("figure", figsize=(8,8))

# define key constants
qe =  -1.6e-19      # charge of electron: C
m = 3.90e-25 #kg, mass of U-235
q = -qe #C, Charge of U-235

dt = 1.0E-8 #s
t = 0 #s starting time

pos = np.array([0,0,0]) #m, sets initial position
v = np.array([0,0,0]) #m/s, sets initial velocity

#create lists for graphing later
xlist = [pos[0]]
ylist = [pos[1]]
zlist = [pos[2]]
tlist = [t]


# Strength of Fields
B = np.array([0,0,0]) #Unit: Teslas, magnetic field is in the z-direction 
E = np.array([50000,0,0]) #unit: N/C, electric field is zero for this example

# Euler time steps
while pos[0] < 0.05:
    Fnet = q*(E + np.cross(v, B)) #Defining net force vector
    a = Fnet / m #calculating accerlation vector, Newton's 2nd Law
    v = v + a*dt #updating velocity vector 
    pos = pos + v*dt #updating position vector
    t += dt #updating time
    
    #updating lists
    xlist.append(pos[0])
    ylist.append(pos[1])
    zlist.append(pos[2])
    tlist.append(t)

print(v[0], "m/s") 

# # Plot the particle's trajectory, in the xy-plane
# plt.plot(xlist,ylist, label="U-235")
# plt.xlabel("$x$")
# plt.ylabel("$y$")
# plt.title("Particle Trajectories in Uniform Magnetic Field")
# plt.legend(fontsize="x-large")
# plt.xlim(-13,13)
# plt.ylim(-13,13)
# plt.show()

# #Now plot the particle's trajector in the xz-plane. What do you notice?
# plt.plot(xlist,zlist, label="Alpha particle")
# plt.xlabel("$x$")
# plt.ylabel("$z$")
# plt.title("Particle Trajectories in Uniform Magnetic Field")
# plt.show()
```

```python
#Part B
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for plot attributes
plt.rc("xtick", labelsize="large")
plt.rc("ytick", labelsize="large")
plt.rc("axes", labelsize="xx-large")
plt.rc("axes", titlesize="xx-large")
plt.rc("figure", figsize=(8,8))

# define key constants
qe =  -1.6e-19      # charge of electron: C
m = 3.90e-25 #kg, mass of U-235
q = -qe #C, Charge of U-235

dt = 1.0E-9 #s
t = 0 #s starting time

# Strength of Fields
B = np.array([0,0,0.441176]) #Unit: Teslas, magnetic field is in the z-direction 
E = np.array([0,0,0]) #unit: N/C, electric field is zero for this example

pos = np.array([0,0,0]) #m, sets initial position
v = np.array([45333.33333333336,0,0]) #m/s, sets initial velocity

#create lists for graphing later
xlist = [pos[0]]
ylist = [pos[1]]
zlist = [pos[2]]
tlist = [t]

#updating lists
xlist.append(pos[0])
ylist.append(pos[1])
zlist.append(pos[2])
tlist.append(t)

# Euler time steps
while t < 1e-4:
    Fnet = q*(E + np.cross(v, B)) #Defining net force vector
    a = Fnet / m #calculating accerlation vector, Newton's 2nd Law
    v = v + a*dt #updating velocity vector 
    pos = pos + v*dt #updating position vector
    t += dt #updating time
    
    #updating lists
    xlist.append(pos[0])
    ylist.append(pos[1])
    zlist.append(pos[2])
    tlist.append(t)

print("radius = ", (max(xlist)-min(xlist))/2, "m")

# Plot the particle's trajectory, in the xy-plane
plt.plot(xlist,ylist, label="U-235")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Particle Trajectories in Uniform Magnetic Field")
plt.legend(fontsize="x-large")
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```
