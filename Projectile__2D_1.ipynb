```python
# put the needed libraries first, here at the top
import matplotlib.pyplot as plt         # "plt" is much shorter
import numpy as np

#Knowns - what variables did the problem give us? don't forget units!
m = 1 # kg, mass
t = 0 #s, Starting time
g = np.array([0, 9.8]) # N/kg gravitational force
pos = np.array([0, 0]) #m, Initial position
v = np.array(np.sqrt([2, 2])) #m/s, velocity
dt = 0.1 #s
g = 9.8 # 9.8 N/kg

#Lists - these will typically be used to create graphs, so think about what we want to graph! Don't forget units
tlist = [t] #s
xlist = [pos[0]] #m
ylist = [pos[1]] #m
vxlist = [v[0]] #m/s
vylist = [v[1]] #m/s

#The Motion - Using a while loop, we update the various variables of the object
while pos[1] >= 0: #example - while t < 10 - read the question closely. Under what condition do we want to the code to stop running?
    Fnet = -m*g # net force
    a = Fnet/m  #solve the acceleration
    v = v + a*dt  #solve the acceleration
    pos = pos + v*dt  #solve the acceleration
    t = t + dt  #solve the acceleration

    #we also need to update our lists
    tlist.append(t)
    xlist.append(pos[0])
    ylist.append(pos[1])
    vxlist.append(v[0])
    vylist.append(v[1])
    print(pos[0], pos[1])

plt.xlabel("x pos (m)")
plt.ylabel("y pos (m)")
plt.title("tragectory graph")

plt.scatter(xlist, ylist, color='pink', edgecolor='blue');   # this makes a scatterplot!

plt.show() 


#Graphs - what are the common physics graphs you'd be asked for a question like this? 
#Do you have a graphing code you can copy and paste here?

#How do we get the computer to tell us the answer? Where is this variable being stored? How do we print it out?
```
